```python
artists.Gender = artists.Gender.str.lower()
artists.Gender.unique()
```

```python
# Make DateAcquired a datetime object
artworks.DateAcquired = pd.to_datetime(artworks.DateAcquired)
```

```python
artists.ConstituentID
```

```python
# N.B. This strips out any works that have multiple artists, which is fine for my purposes
artworks = artworks[artworks.ConstituentID.str.contains(",") == False]
```

```python
''' We have to make the ConstituentID series in the artworks dataframe integers, 
now that we've dropped the rows that have multiple IDs separated by a comma'''

artworks.loc[:, 'ConstituentID'] = artworks.loc[:, 'ConstituentID'].astype(int)
```

```python
# Create a dataset that combines artists and artworks according to ConstituentID
df = pd.merge(artworks, artists, on="ConstituentID")
```

```python
# Make gender lowercase to normalize
df.Gender_y = df.Gender_y.str.lower()
```

```python
df.Gender_y.unique()
```

```python
# N.B. This strips out any works that not associated with a gender, which is fine for my purposes
df = df[df.Gender_y.notna()]
```

```python
df.head()
```

```python
df.columns
```

```python
# Create a new dataframe with just the subset of the columns we need

ff = df[['Title','Artist','DisplayName','ConstituentID','ArtistBio_y', 'Nationality_y',
       'Gender_y', 'BeginDate_y', 'EndDate_y', 'Wiki QID', 'ULAN', 'Date', 'Medium', 'Dimensions',
       'CreditLine', 'AccessionNumber', 'Classification', 'Department',
       'DateAcquired', 'Cataloged', 'ObjectID', 'URL', 'ThumbnailURL']]
```

```python
ff.loc[:, 'DateAcquired'] = pd.to_datetime(ff.DateAcquired)
```

```python
# Create a new dataframe that counts the object per gender per year
gg = ff.groupby([ff.DateAcquired.dt.year, ff.Gender_y]).count()
gg = gg.rename_axis(['DateAcquired_yr', 'Gender'])
gg = gg.reset_index()
gg_out = gg[['DateAcquired_yr','Gender','Title']]
gg_out = gg_out.rename(columns={"Title": "Count"})
gg_out.to_csv("moma_data.csv")
```

```python
gg_out
```

```python
# Reshape the data to make it a little easier to work with by default D3
pivoted = gg_out.pivot(index="DateAcquired_yr", columns="Gender", values="Count")
pivoted.to_csv("moma_data_shaped.csv")
pivoted
```

## Trying unique artists per year instead of total artworks per year based on class feedback

```python
# Group by artists per year, take the first row of data in each grouping to stand in for now
u = ff.groupby([ff.DateAcquired.dt.year, ff.ConstituentID]).first()
```

```python
# Now group by year and by gender, and note counts by grouping
u = u.groupby([u.DateAcquired.dt.year, u.Gender_y]).count()
```

```python
# Create a dataframe of the unique artists per year by gender
u = u.rename_axis(['DateAcquired_yr', 'Gender'])
u = u.reset_index()
uniques = u[['DateAcquired_yr','Gender','Title']]
uniques = uniques.rename(columns={"Title": "Count"})
uniques.to_csv("unique_artists_moma_data.csv")
```

```python
uniques
```

```python
# Reshape the data to make it a little easier to work with by default D3
pivoted_uniques = uniques.pivot(index="DateAcquired_yr", columns="Gender", values="Count")
pivoted_uniques.to_csv("unique_artists_moma_data_shaped.csv")
```

```python
pivoted_uniques
```

## Now do unique artists per year by gender, but also add in Classifications

```python
v = ff.groupby([ff.DateAcquired.dt.year, ff.ConstituentID, ff.Classification]).first()
v = v.rename_axis(['DateAcquired_yr', 'Gender', 'Classification'])
v = v.reset_index()
v = v.groupby([v.DateAcquired.dt.year, v.Gender_y, v.Classification])["Title"].count().reset_index(name="gender_year_classification_count")
v.loc[:, 'gender_year_count'] = v.groupby(["DateAcquired", "Gender_y"])['gender_year_classification_count'].transform(sum)
```

```python
v
```

```python
# Add a row for Classification: "All" as an affordance, so in our initial D3 view we don't have to do any calculations
v = v.set_index(['DateAcquired','Gender_y', "Classification"])
for row in v.index:
    year = row[0]
    gender = row[1]
    classification = row[2]

    v.at[(year, gender, "All"), "gender_year_classification_count"] = v.loc[row, "gender_year_count"]
```

```python
# Cleanup and rename columns to more useful names
v = v[['gender_year_classification_count']]
v = v.reset_index()
v = v.rename(columns={"gender_year_classification_count": "count", 
                      "DateAcquired" : "year",
                      "Gender_y" : "gender", 
                     "Classification" : "classification"})
```

```python
v = v.sort_values(by=['year','gender','classification'], ignore_index=True, ascending=[True, True, True])
```

```python
v
```

```python
# reshape the data, once again, to make it more easily useful for D3
v = v.pivot(index=["year", "classification"], columns="gender", values="count").fillna(0)
```

```python
v.to_csv('moma_data_shaped_with_classification.csv')
```
