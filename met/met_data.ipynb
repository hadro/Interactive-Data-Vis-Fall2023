```python
artists.Gender = artists.Gender.str.lower()
artists.Gender.unique()
```

```python
artworks.DateAcquired = pd.to_datetime(artworks.DateAcquired)
```

```python
artworks.Title.groupby(artworks.DateAcquired.dt.year).count()
```

```python
artists.ConstituentID
```

```python
artworks = artworks[artworks.ConstituentID.str.contains(",") == False]
```

```python
artworks.ConstituentID = artworks.ConstituentID.astype(int)
```

```python
df = pd.merge(artworks, artists, on="ConstituentID")
```

```python
df.Gender_y = df.Gender_y.str.lower()
```

```python
df.Gender_y.unique()
```

```python
df = df[df.Gender_y.notna()]
```

```python
df.head()
```

```python
df.columns
```

```python
ff = df[['Title','Artist','DisplayName','ConstituentID','ArtistBio_y', 'Nationality_y',
       'Gender_y', 'BeginDate_y', 'EndDate_y', 'Wiki QID', 'ULAN', 'Date', 'Medium', 'Dimensions',
       'CreditLine', 'AccessionNumber', 'Classification', 'Department',
       'DateAcquired', 'Cataloged', 'ObjectID', 'URL', 'ThumbnailURL']]
```

```python
ff.DateAcquired = pd.to_datetime(ff.DateAcquired)
```

```python
gg = ff.groupby([ff.DateAcquired.dt.year, ff.Gender_y]).count()
```

```python
gg = gg.rename_axis(['DateAcquired_yr', 'Gender'])
```

```python
gg = gg.reset_index()
```

```python
out = gg[['DateAcquired_yr','Gender','Title']]
```

```python
out = out.rename(columns={"Title": "Count"})
```

```python
out.to_csv("met/met_data.csv")
```

```python
out
```

```python
pivoted = out.pivot(index="DateAcquired_yr", columns="Gender", values="Count")
```

```python
pivoted.to_csv("met/met_data_shaped.csv")
```
